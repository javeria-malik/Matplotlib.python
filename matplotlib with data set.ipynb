# color

```python
plt.plot(data["index"],data["V Kohli"],color="green")
plt.plot(data["index"],data["RG Sharma"],color="black")
plt.title("comparison")
plt.xlabel("year")
plt.ylabel("batsman career")
```

# style line

```python
plt.plot(data["index"],data["V Kohli"],color="green",linestyle="dashdot")
plt.plot(data["index"],data["RG Sharma"],color="black",linestyle="solid")
plt.title("comparison")
plt.xlabel("year")
plt.ylabel("batsman career")
```

# can aslo work with length of line

```python
# sine the point is called marker so we cam also highlight them by parametetr called marker
```

```python
plt.plot(data["index"],data["V Kohli"],color="green",linestyle="dashdot",marker="D",label="virat")# for this it reprent d
plt.plot(data["index"],data["RG Sharma"],color="black",linestyle="solid",marker="o",markersize=10,label="sharma")#for this lets have another
plt.title("comparison")
plt.xlabel("year")
plt.ylabel("batsman career")
plt.legend(loc="upper right")# cam also give right left parameters
```

```python
plt.bar(data2["batsman"],data2["2015"],width=0.3)
```

```python
plt.bar(np.arange(data2.shape[0])-0.5,data2["2015"],width=0.75)#will give numbers instead of name
```
